# Feature Engineering

```python
data.specialization_higher_edu.unique()
```

```python
data.field_degree.unique()
```

```python
data.post_grad_specia.unique()
```

```python
# Encoding categorical data

data.gender = data.gender.map({'M': 0, 'F': 1})
data.post_grad_specia = data.post_grad_specia.map({'Mkt&HR': 0, 'Mkt&Fin': 1})
data.status = data.status.map({'Placed': 1, 'Not Placed': 0})
```

```python
# dropping cols
data.drop(['Secondary_education_board', 'higher_edu_board', 'specialization_higher_edu'], axis=1, inplace=True)
```

```python
# Encoding the categorical data
data = pd.get_dummies(data, columns=['field_degree', 'work_exp'], drop_first=True)
```

```python
data.head()
```

```python
# filling missing values
data.salary = data.salary.fillna(data.salary.mean())
```

```python
data.isnull().sum()
```

# Splitting data

```python
X = data.drop('status', axis=1)
y = data.status
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

# Creating Model and checking accuracy

```python
def trees(X_train, X_test, y_train, y_test):
    from sklearn.tree import DecisionTreeClassifier
    from sklearn.metrics import mean_squared_error
    from sklearn.metrics import confusion_matrix
    model = DecisionTreeClassifier()
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    return mean_squared_error(y_test, preds), confusion_matrix(y_test, preds)
```

```python
print(trees(X_train, X_test, y_train, y_test))
```
